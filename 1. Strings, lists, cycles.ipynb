```python
#Найти наименьший нечетный

numbers = list(map(int, input().split()))
m = 0
for number in numbers:
    if number % 2 != 0 and (m > number or m == 0):
        m = number
print(m)
```

```python
# На ввод даны числа a и b. Нужно вывести все целые числа в промежутке от a до b включительно . Если a<b, выводим по возрастанию, если наоборот - по убыванию.

a = int(input())
b = int(input())
if a < b:
    for i in range(a, b + 1):
        print(i)
else:
    for i in range(a, b - 1, -1):
        print(i)
```

```python
# Сравнивать пары чисел (и считать в скольких случаях следующее введенное число больше предыдущего) до тех пор, пока не будет введен ноль. 

prev_number = int(input())
cur_number = int(input())
cnt = 0
while cur_number != 0:
    if cur_number > prev_number:
        cnt += 1
    prev_number = cur_number
    cur_number = int(input())
print(cnt)
```

```python
# Сравнивать пары чисел (и считать в скольких случаях следующее введенное число больше предыдущего).
# На вход поступает число N - количество сравниваемых чисел. Затем вводятся сравниваемые числа. 

N = int(input())
prev_num = int(input())
cnt = 0
for i in range(N - 1):
    cur_num = int(input())
    if cur_num > prev_num:
        cnt += 1
    prev_num = cur_num
print(cnt)
```

```python
#Необходимо найти первые встретившиеся соседние элементы одного знака

num = list(map(int, input().split()))
num_prev = num[0]
for i in range(1, len(num)):
    if num_prev * num[i] > 0:
        print(num_prev, num[i])
        break
    else:
        num_prev = num[i]
```

```python
# КЕГЛИ
# На вход поступают N - количество кеглей, K - количество бросков
# Далее вводятся пары чисел - промежуток, между которым сбиты кегли (включая введенные номера)
# Ответ - напечатать итоговый ряд кеглей (I - остались не сбитыми, . - сбиты)

N, K = list(map(int, input().split()))
bowls_row = ['I'] * N
for i in range(K):
    l_i, r_i = list(map(int, input().split()))
    l_i -= 1
    r_i -= 1
    for j in range(l_i, r_i + 1):
        bowls_row[j] = '.'
print(*bowls_row, sep='')
```

```python
# Сортировка списка , вариант 1 (неоптимальный)
# Дан список чисел (любого размера) со значениями чисел от 0 до 100. Необходимо отсортировать его по возрастанию.

lst = list(map(int, input().split()))
count_lst = []


for elem in list(range(101)):
    if elem in lst:
        x = lst.count(elem) #невыгодная операция (каждый раз заново проходится по списку)
        for i in range(x):
            count_lst.append(elem)     
print(count_lst)
```

```python
#Сортировка списка, вариант 2 - сортировка подсчетом (более оптимальный за счет одного прохода по двум спискам)
# Дан список чисел (любого размера) со значениями чисел от 0 до 100. Необходимо отсортировать его по возрастанию.

lst = list(map(int, input().split()))
count_lst = [0] * 101 # последний индекс 100

for i in lst:
    count_lst[i] += 1

for i in range(len(count_lst)):
    for num in range(count_lst[i]): #в случае count_lst[i] == 0 просто не заходим во вложенный цикл
        print(i, end=' ')
```

```python
#Сортировка списка - вариант 3 - сортировка подсчетом (более короткий, без вложенного цикла для печати итогового списка)
# Дан список чисел (любого размера) со значениями чисел от 0 до 100. Необходимо отсортировать его по возрастанию.

lst = list(map(int, input().split()))
count_lst = [0] * 101

for i in lst:
    count_lst[i] += 1

for i in range(len(count_lst)):
    print((str(i) + ' ') * count_lst[i], end='')
```

#### КПП


```python
def avg_age(age_lst):
    return int(sum(age_lst)/len(age_lst))

name_lst = []
age_lst = []
clients = int(input())
for i in range(clients):
    name = input('What is your name?')
    age = int(input('How old are you?'))
    if age >= 18:
        name_lst.append(name)
        age_lst.append(age)
        print('YES')
    else: 
        print('NO')
passed = len(name_lst)
if passed == 0:
    print ('Passed: ' + str(passed))
else:
    print('Passed: ' + str(passed))
    print('Minimal age: ' + str(min(age_lst)))
    print('Maximal age: ' + str(max(age_lst)))
    print('Average age: '+ str(avg_age(age_lst)))
    
    name_len = 0
    for i in range(len(name_lst)):
        name_len += len(name_lst[i])
    print('Average name length: ' + str((name_len / len(name_lst))))
```

#### Сдвиг строк


```python
s_k = input()
s_d = input()
x = len(s_k)
for i in range(len(s_k)):
    if s_k == s_d:
        print(0)
        break
    elif s_k[i:] + s_k[:i] == s_d:
        print(x)
        break
    else:
        x -= 1
else:
    print(-1)
```

#### Палиндромы


```python
st = input()
lst = list(st)

if lst == lst[::-1]:
    print ('YES')
else:
    for i in range(len(lst)):
        lst_new = lst[:i] + lst[1 + i:] 
        if lst_new == lst_new[::-1]:
            print ('YES')
            break
    else:
        print('NO')
```

```python
n = int(input())
lst = [[0 for j in range(n)] for i in range(n)]
for i in range(n):
        lst[i][n - i - 1] = 1
for i in range(1, n):
    for j in range(n):
        if lst[i][j] != 0 and j + 1 < n:
            lst[i][j + 1] = 2
for row in lst:
    for elem in row:
        print(elem, end=' ')
    print()
```
