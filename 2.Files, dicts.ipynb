```python
# Посчитать сколько каждое слово встречается в предложении и вывести значения для каждого слова в столбик

seq = input().split()
count_dct = {}


for word in seq:
    count_dct[word] = count_dct.get(word, 0) + 1

for key in count_dct:
    print(key, ': ', count_dct[key], sep='')
```

```python
# Посчитать сколько раз каждое слово встречается в текстовом файле

inFile = open('input.txt', encoding='utf-8')
dct = {}
for line in inFile:
    words = line.rstrip().split()
    for word in words:
        #dct[word] = dct.get(word, 0)
        if word not in dct:
            dct[word] = 0
        dct[word] += 1

inFile.close()

for word in dct:
    print(word, ': ', dct[word], sep='')
```

```python
# Cтраны и города
# На ввод поступает количество стран, далее последовательно вводятся списки, где первой идет страна, затем города в ней
# Далее вводится количество городов и последовательно названия каждого города
# Необходимо определить к какой стране принадлежит город

N = int(input())
dct = {}

for i in range(N):
    lst = input().split()
    country = lst[0]
    towns = lst[1:]
    for town in towns:
        dct[town] = country
M = int(input())
for i in range(M):
    town = input()
    country = dct[town]
    print(country)
```

#### Первый уникальный символ

```python
string1 = input()
string2 = string1.split()

letters = []
letters_dct = {}

for elem in string2:
    for letter in list(elem):
        letters.append(letter)

for letter in letters:
    letters_dct[letter] = letters_dct.get(letter, 0) + 1

for letter in letters:
    if letters_dct[letter] == 1:
        print(list(string1).index(letter))
        break
else:
    print('NOTFOUND')
```

#### Поисковые запросы 4

```python

with open('input.txt', encoding='utf-8') as f:
    lines = f.read().splitlines()

k = int(lines.pop(0)) #в условии сказано, что 'Первая строка входных данных содержит число k.', поэтому здесь ее удаляю
pages_dct = {}
uniq_queries = {}

for string in lines:
    ts, dt, device, pageid, regid, url, links = string.split('\t')
    
    links_all = links.split(';')
    
    assert url.startswith('https://yandex.ru/search/?text=')
    query_and_other_params = url[31:]

    if '&' in query_and_other_params:
        query = query_and_other_params[:query_and_other_params.index('&')]
    else:
        query = query_and_other_params
    uniq_queries[query] = links_all
    
    for link in links_all:
        pages_dct[link] = pages_dct.get(link, 0) + 1

n = set()
for page in pages_dct:
    if pages_dct[page] > k:
        for query in uniq_queries:
            if page in uniq_queries[query]:
                n.add(query)
print(len(n)) 
```
